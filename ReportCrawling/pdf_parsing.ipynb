```python
data = pd.DataFrame()
for foldername in tqdm(os.listdir("html_files")):
    file = []
    try : 
        for filename in os.listdir('./html_files/'+foldername):
            if filename.endswith(".html"):
                file_dir = '/Users/sera/company/ReportCrawling/html_files/'+foldername+'/'+filename
                page = open(file_dir,'rt',encoding='utf-8').read()
                bs = BeautifulSoup(page,'html.parser')
                file.append(bs.prettify())
        tmp = pd.DataFrame(zip([foldername],[file]))
        data = pd.concat([data,tmp],ignore_index=True)
    except : 
        error.append(foldername)
data.columns = ['filename','content']       
```

```python
data = pd.read_csv('reportdata.csv')
data.content = list(map(lambda x : eval(x), data.content))
```

```python
%time
import re
headers =[]
for i in range(len(data)):
    content=[]
    for j in range(len(data.content[i])):
        tmp = BeautifulSoup(data.content[i][j],\
                            'html.parser').findAll('span',\
                                                   {'style':[re.compile("font-size:[\d]+px;vertical-align:baseline;color:rgba\(0,160,198,1\);"),
                                                             re.compile("font-size:[\d]+px;vertical-align:baseline;color:rgba\(0,159,197,1\);"),
                                                             re.compile("font-size:[\d]+px;vertical-align:baseline;color:rgba\(85,146,206,1\);"),
                                                             re.compile("font-size:[\d]+px;vertical-align:baseline;color:rgba\(85,145,205,1\);"),
                                                             re.compile("font-size:[\d]+px;vertical-align:baseline;color:rgba\(63,63,63,1\);")
                                                                        ]})
        content.extend(tmp)
        
        
    if content == []:
        headers.append(content)
        print(i)
    else : 
        headers.append(content)
```
